```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
fig, ax = plt.subplots()
x = np.linspace(-0.5, 0.5, 100)
for s in np.linspace(0, 1, 5):
    ax.plot(x, norm.pdf(x, scale=s), label='stdv=%.1f' % s)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.grid(True)
```

```python
def fn(x, scale):
    c = 1.0 - (scale * 0.5)
    for i in range(x.shape[0]):
        x[i] = c
        c *= scale
    return x

fig, ax = plt.subplots()
x = np.linspace(0, 10, 10)
for s in np.linspace(1, 0, 5):
    ax.plot(x, fn(x.copy(), scale=s), label="Curve (%.2f)" % s)
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Probability of words')
ax.legend(loc='best', frameon=True)
ax.grid(True)
```
