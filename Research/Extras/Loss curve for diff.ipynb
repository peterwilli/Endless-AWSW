```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
```

```python
x = np.linspace(0, 0.0005, 10)
weights = [
    i ** 3 for i in range(0, 10)
]
y = np.array([w for w in weights])
f = interpolate.interp1d(x, y, kind = 'quadratic', fill_value = 'extrapolate')
fig, ax = plt.subplots()
x = np.linspace(0, 0.0005, 100)
xf = [f(xp) for xp in x]
ax.plot(x, xf, label="Curve")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Loss for diff')
ax.legend(loc='best', frameon=True)
ax.grid(True)
```
